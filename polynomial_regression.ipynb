# polynomial_regression


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("polynomial.csv")
data.head(5)
```

```python
plt.scatter(data["level"],data["salary"])
plt.xlabel("salary",fontsize="20",color="red")
plt.ylabel("level",fontsize="20",color="red")
plt.show()
```

```python
input_a=data[["level"]]
output_b=data["salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2)
pf.fit(input_a)
x=pf.transform(input_a)
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_a,output_b,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
prd=lr.predict(input_a)
prd
```

```python
plt.scatter(data["level"],data["salary"])
plt.plot(data["level"],prd,color="red")
plt.xlabel("salary",fontsize="20",color="red")
plt.ylabel("level",fontsize="20",color="red")
plt.show()
```
