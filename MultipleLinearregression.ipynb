# MultipleLinearRegression


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("agesalaryExp.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
int_a=dataset.iloc[:,:-1]
out_b=dataset["Salary"]
out_b
```

```python
sns.pairplot(data=dataset.head(30))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(int_a,out_b,test_size=0.20,random_state=50)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.predict(x_test)
```
