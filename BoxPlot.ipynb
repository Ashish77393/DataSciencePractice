# BoxPlot

```python
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40,50,60,120]
y=[10,20,10,40,50,60,80]
c=[23,44,33,22,55,66,88]
z=[x,y,c]
plt.boxplot(z,notch=False,vert=True,widths=0.1,labels=["python","java","c++"],patch_artist=False,showmeans=True,whis=1.5,sym="r+",capprops=dict(color="r"),
           boxprops=dict(color="green"),whiskerprops=dict(color="orange"))
```
