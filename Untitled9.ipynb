```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
y_1=sns.load_dataset("penguins")
y_1
order_1=["Dream","Torgersen","Biscoe"]

sns.barplot(x="island",y="bill_length_mm",data=y_1,hue="sex",order=order_1,saturation=10,errcolor="g",dodge=True,alpha=0.5)
plt.grid(color="red",alpha=0.3)
sns.set(style="darkgrid")

plt.show()
```
