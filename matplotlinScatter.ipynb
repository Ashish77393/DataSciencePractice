# Scatter Graph Bar

```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4,5,6,7]
y=[3,4,3,5,6,5,6]
z=[3,4,3,4,6,4,5]
plt.title("Scatter Graph",fontsize=20,color="red",rotation=10)
plt.xlabel("Number of Student",fontsize=20,color="green")
plt.ylabel("no of course",fontsize=20,color="green")
color=["red","green","blue","pink","violet","indigo","orange"]
size=[200,300,400,500,200,300,400]
linewidth=[1,2,1,3,2,1,3]
color1=[45,88,55,33,55,66,77]
plt.scatter(x,y,c=color,s=size,alpha=0.9,marker="*",edgecolor="green",linewidth=0,cmap="Accent",label="WSCUBE")
plt.scatter(x,z,color="red",s=size,alpha=0.9,edgecolor="green",linewidth=0,cmap="Accent")
t=plt.colorbar()
t.set_label("colorBar",fontsize=30,color="orange")
```
