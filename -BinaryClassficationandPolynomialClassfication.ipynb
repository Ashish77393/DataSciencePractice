```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\Ashish Kumar Singh\Downloads\datasetlinearRegression.csv")
dataset.head(3)
```

```python
sns.scatterplot(x="Age",y="salary",data=dataset,hue="Purchased")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf= PolynomialFeatures(degree=6)
pf.fit(x)
newdataset=pd.DataFrame(pf.transform(x))
newdataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from mlxtend .plotting import plot_decision_regions
```

```python
 plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
 plt.show()
```
