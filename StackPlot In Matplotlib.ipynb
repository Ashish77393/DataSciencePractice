# Stack Plot

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
area1=[2,1,3,5,3,2]
area2=[3,5,2,1,2,1]
area3=[1,4,2,6,3,2]
area4=[3,2,1,5,3,2]
l=["area1","area2","area3","area4"]
c=['r','g','b','orange']
plt.stackplot(x,area1,area2,area3,area4,labels=l,colors=c,baseline="zero")
plt.legend()
plt.title("Stack Plot graph",fontsize=15,color="red")
plt.xlabel("x-axis",fontsize=15,color="red")
plt.ylabel("y-axis",fontsize=15,color="red")
# plt.grid(color="red")
```
