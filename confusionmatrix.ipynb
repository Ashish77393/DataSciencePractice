```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("student_placed.csv")
dataset.head(3)
```

```python
input=dataset.iloc[:,:-1]
output=dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input,output,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score
```

```python
cf=confusion_matrix(y_test,lr.predict(x_test))
```

# graph of cunfusion mconfusion_matrix

```python
plt.figure(figsize=(3,2))
sns.heatmap(cf,annot=True)
plt.show()
```

```python
recall_score(y_test,lr.predict(x_test))*100
```

```python
f1_score(y_test,lr.predict(x_test))*100
```
