```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6]
y=[2,4,2,6,4,7]
```

```python
plt.figure(figsize=(5,4))
plt.bar(x,y,width=0.4,label="Ashish")
plt.xlabel("Language",fontsize=20)
plt.ylabel("Popolation Speak",fontsize=20)
plt.title("graph of people speak language",color="red")
plt.legend()
plt.show()
```
