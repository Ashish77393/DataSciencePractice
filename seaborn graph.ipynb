# histogram plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("penguins").head(20)
df
sns.displot(df["bill_length_mm"],kde=True,rug=True,color="green")
plt.grid()
plt.show()
```

# Scatter plot


```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",size="sex",sizes=(100,200),alpha=0.5)
plt.show()
```

# countplot

```python
var=sns.load_dataset("tips")
var
sns.countplot(x="sex",data=var,hue="smoker",saturation=0.9,alpha=0.7)
plt.show()
```

# Violin plot

```python
sns.violinplot(x="time",y="total_bill",data=var,hue="time",linewidth=0.1,order=["Dinner","Lunch"],inner="point",dodge=True)
plt.show()
```
