# labelEncoding on ordinalData

```python
dataset=pd.DataFrame({
    "size":['s','s','m','l','s','m','l','s','l','s','l','xl','xl','s','m','xl','s','l','s','xl']
})
dataset
```

```python
from sklearn.preprocessing import OrdinalEncoder 
```

```python
sizelist=[['s','m','l','xl']]
```

```python
oe=OrdinalEncoder(categories=sizelist)
dataset["oe_size"]=oe.fit_transform(dataset[['size']])
```

```python
dataset
```

# using Map Function

```python
ord_data_enc={
    "s":0,
    "m":1,
    "l":2,
    "xl":3
}
```

```python
dataset['map_size_encoder']=dataset["size"].map(ord_data_enc)
```
