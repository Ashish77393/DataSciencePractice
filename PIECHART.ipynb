# PIE CHART

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=['c','c++','java','python']
z=[30,40,10,20]
ex=[0.1,0.0,0.0,0.0]
c=['red','green','orange','grey']
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",radius=1.5,labeldistance=1.1,startangle=90,textprops={"fontsize":15},
       counterclock=False,rotatelabels=True)
plt.title("Popularity of Language")
plt.legend(loc=3)
```
