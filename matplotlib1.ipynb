# BAR PLOT

```python
x=['c','c++','java','python','js']
y=[70,72,82,85,80]
z=[22,44,33,55,66]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=20,color="green")
plt.ylabel("NO",fontsize=20,color="green")
plt.title("graph of language",fontsize=25,color='red')
plt.bar(p,y,color='r',width=0.2,label="Popularity of language")
plt.bar(p1,z,width,color='yellow',label="popularity of language1")
plt.xticks(p+width/2,x,rotation=10)
plt.legend()

```

```python
plt.xlabel("language",fontsize=20,color="green")
plt.ylabel("NO",fontsize=20,color="green")
plt.title("graph of language",fontsize=25,color='red')
plt.barh(p,y,color='r',width=0.2,label="Popularity of language")
plt.barh(p1,z,width,color='yellow',label="popularity of language1")
plt.xticks(p+width/2,x,rotation=10)
plt.legend()
```
