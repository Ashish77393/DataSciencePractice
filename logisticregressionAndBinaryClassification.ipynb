Logistic Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("datasetlinearRegression.csv")
```

```python
dataset.head(3)
```

```python
dataset.drop(columns=["salary"],inplace=True)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="Age",y="Purchased",data=dataset)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_x=dataset[["Age"]]
output_y=dataset["Purchased"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_x,output_y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[46]])
```

```python
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.lineplot(x="Age",y=lr.predict(input_x),data=dataset)
plt.show()
```
