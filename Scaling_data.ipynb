# Sacaling_Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv("Mobile_sales.csv")
data.head(3)
```

```python
data.describe()
```

```python
sns.distplot(data["TotalRevenue"])
plt.show()
```

```python
ns=MinMaxScaler()
ns.fit(data[["TotalRevenue"]])
```

```python
ns.transform(data[["TotalRevenue"]])
```

```python
data["TotalRevenue_ns"]=ns.transform(data[["TotalRevenue"]])
```

```python
data
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.title("AFter Graph")
sns.distplot(data["TotalRevenue_ns"])

plt.subplot(1,2,2)
plt.title("Before Graph")
sns.distplot(data["TotalRevenue"])
```
